```python
# 1. Installatie van vereiste bibliotheken
!pip install open3d

import open3d as o3d
import numpy as np

def generate_pcd():
    pts = np.random.rand(1000, 3)
    pcd = o3d.geometry.PointCloud()
    pcd.points = o3d.utility.Vector3dVector(pts)
    return pcd

# Voxel downsampling (Voxel size: 0.05)
pcd = generate_pcd()
down = pcd.voxel_down_sample(voxel_size=0.05)

print(f"Punten na reductie: {len(down.points)}")

```

```python
import plotly.graph_objects as go

# Get points from the original point cloud
pcd_points = np.asarray(pcd.points)

# Create a Plotly 3D scatter plot for the original point cloud
fig_original = go.Figure(
    data=[go.Scatter3d(
        x=pcd_points[:, 0],
        y=pcd_points[:, 1],
        z=pcd_points[:, 2],
        mode='markers',
        marker=dict(
            size=2,
            opacity=0.8
        )
    )],
    layout=go.Layout(
        margin=dict(l=0, r=0, b=0, t=0),
        title_text='Originele Puntwolk'
    )
)
fig_original.show()

# Get points from the downsampled point cloud
down_points = np.asarray(down.points)

# Create a Plotly 3D scatter plot for the downsampled point cloud
fig_downsampled = go.Figure(
    data=[go.Scatter3d(
        x=down_points[:, 0],
        y=down_points[:, 1],
        z=down_points[:, 2],
        mode='markers',
        marker=dict(
            size=2,
            opacity=0.8
        )
    )],
    layout=go.Layout(
        margin=dict(l=0, r=0, b=0, t=0),
        title_text='Gereduceerde Puntwolk'
    )
)
fig_downsampled.show()
```

```python
output_path_original = "original_point_cloud.ply"
o3d.io.write_point_cloud(output_path_original, pcd)
print(f"Originele puntwolk opgeslagen als: {output_path_original}")

output_path_downsampled = "downsampled_point_cloud.ply"
o3d.io.write_point_cloud(output_path_downsampled, down)
print(f"Gereduceerde puntwolk opgeslagen als: {output_path_downsampled}")
```
