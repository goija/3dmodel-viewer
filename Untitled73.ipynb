```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.animation import FFMpegWriter
from matplotlib.collections import LineCollection
from tqdm import tqdm

def create_hex_video_verbeterd(scale=16666.67, divisions=100, rows=3, cols=3, filename="tesselatie_G_final.mp4"):
    fig, ax = plt.subplots(figsize=(10, 10), dpi=100)

    # Gebruik de schaal voor de zoom-factor
    R_base = 1.0
    dx_base, dy_base = 1.5 * R_base, np.sqrt(3) * R_base

    writer = FFMpegWriter(fps=30, metadata=dict(artist='mevrouw G.'), bitrate=4000)

    with writer.saving(fig, filename, 100):
        for frame in tqdm(range(60), desc="Rendering"):
            ax.clear()
            ax.set_facecolor('#f5f2e9') # Uw gekozen achtergrondkleur

            # Oscillatie en Zoom (gebaseerd op uw reeks 50k - 500k)
            shift = (np.sin(2 * np.pi * frame / 60) + 1) / 2
            zoom = 1 + (shift * 0.2) # Subtiele zoom-beweging

            color = plt.cm.viridis(frame / 60)
            all_lines = []

            for r in range(rows):
                for c in range(cols):
                    cx = c * dx_base
                    cy = (r * dy_base) + (dy_base / 2 if c % 2 == 1 else 0)
                    center = np.array([cx, cy])

                    angles = np.linspace(0, 2 * np.pi, 7)
                    hex_pts = [center + np.array([np.cos(a), np.sin(a)]) * zoom for a in angles]

                    for i in range(6):
                        p1, p2 = hex_pts[i], hex_pts[i + 1]
                        for n in range(divisions + 1):
                            t = (n / divisions) * shift
                            start = center + t * (p1 - center)
                            end = p1 + (1 - t) * (p2 - p1)
                            all_lines.append([start, end])

            # Teken alle lijnen in één keer voor snelheid
            lc = LineCollection(all_lines, colors=color, linewidths=0.6, alpha=0.5)
            ax.add_collection(lc)

            ax.set_xlim(-1, 5)
            ax.set_ylim(-1, 6)
            ax.axis('off')

            writer.grab_frame() # Nu op de juiste plek

# Voorbeeld aanroep met de hoge resolutie noemer (100)
# create_hex_video_verbeterd(divisions=100)
```
